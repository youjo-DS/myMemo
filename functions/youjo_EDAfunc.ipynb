関数にデータフレームをぶちこむと、各列のデータ型、ユニーク数、ユニーク割合、ユニーク名、欠損数、欠損割合を算出する関数を作ってみた

```python
import pandas as pd
import numpy as np
```

```python
def get_info(df):
    pd.set_option('display.max_rows', None)
    
    cols = df.columns
    
    types = np.array(df.dtypes)
    unique_cnt = df.nunique()
    unique = [df[col].unique() for col in cols]
    null_cnt = np.array(df.isnull().sum())
    
    columns = ['dtypes', 'unique_cnt','unique', 'null_cnt']
    index = df.columns

    disp = pd.DataFrame(np.c_[types, unique_cnt, unique, null_cnt],
                        columns=columns,
                        index=index)
    disp['unique%'] = (disp['unique_cnt'] / len(df)) * 100
    disp['null%'] = (disp['null_cnt'] / len(df)) * 100
    disp = disp[['dtypes', 'unique_cnt', 'unique%', 'unique', 'null_cnt', 'null%']]
    
    print('DataFrame length: ', len(df))
    display(disp)
    
    pd.reset_option('display.max_rows')
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
get_info(df)
```
